**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Don't be afraid to use StackOverflow for help  with this.

```python
# Step 1: Transpose the original DataFrame
properties_T = properties.transpose()

# Step 2: Set the first row as column headers
properties_T.columns = properties_T.iloc[0]  # set first row as header
properties_T = properties_T[1:]              # drop the header row

# Step 3: Reset index so the date becomes a column
properties_T.reset_index(inplace=True)

# Step 4: Rename the index column to 'Date'
properties_T.rename(columns={'index': 'Date'}, inplace=True)

# Step 5: (Optional) Rename first two columns if needed
# You’ll now see columns like:
# Date | City of London | Barking & Dagenham | ...

print(properties_T.columns[:10])  # check first few columns
properties_T.head()        # preview cleaned DataFrame
```

**2.3. Cleaning the data] (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# We're using list comprehension to handle dynamic naming (e.g., NaT)
properties_T.columns = properties_T.columns.tolist()  # ensure it's mutable

# Rename the first three columns manually
new_columns = properties_T.columns.tolist()
new_columns[0] = 'City'
new_columns[1] = 'Code'
properties_T.columns = new_columns


# Step 3: Check the result
print(properties_T.columns[:10])
properties_T.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Melt the DataFrame: keep 'City' and 'Code' as identifiers
properties_melted = pd.melt(properties_T, 
                             id_vars=['City', 'Code'], 
                             var_name='Date', 
                             value_name='AveragePrice')

# Convert Date to datetime format
properties_melted['Date'] = pd.to_datetime(properties_melted['Date'])

# Preview result
properties_melted.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Remove commas, currency symbols, or anything non-numeric (if present)
properties_melted['AveragePrice'] = (
    properties_melted['AveragePrice']
    .astype(str)                      # ensure string for replacement
    .str.replace(',', '')             # remove commas
    .str.replace('£', '')             # remove pound symbol if present
    .str.strip()                      # remove leading/trailing spaces
)

# Convert to float
properties_melted['AveragePrice'] = pd.to_numeric(properties_melted['AveragePrice'], errors='coerce')

# Optional: Check if any rows failed to convert
print(properties_melted['AveragePrice'].isna().sum(), "non-numeric rows found (converted to NaN)")

# Confirm dtype
properties_melted.dtypes
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
print(properties_melted.count())
# Unique values in the London Borough (City) column
print("Unique Boroughs:", properties_melted['City'].nunique())
print(properties_melted['City'].unique())
# Drop rows where 'City' or 'AveragePrice' is null
properties_cleaned = properties_melted.dropna(subset=['City', 'AveragePrice'])

# Optional: Drop non-borough rows (e.g. using a whitelist of official boroughs)
# If you want, I can give you the official list of the 32 boroughs for exact filtering.
print("Final shape:", properties_cleaned.shape)
print("Final unique boroughs:", properties_cleaned['City'].nunique())
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# Pick your borough — change this to any you like
borough_name = "Camden"

# Filter the dataset
borough_data = properties_cleaned[properties_cleaned['City'] == borough_name]
plt.figure(figsize=(12, 6))
plt.plot(borough_data['Date'], borough_data['AveragePrice'])
plt.title(f'Average House Price in {borough_name} Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price (£)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:
1. look through the `Month` column
2. extract the year from each individual value in that column 
3. store that corresponding year as separate column. 

Whether you go ahead with this is up to you. Just so long as you answer our initial brief: which boroughs of London have seen the greatest house price increase, on average, over the past two decades? 

```python
properties_cleaned['Year'] = properties_cleaned['Date'].apply(lambda x: x.year)
yearly_avg = properties_cleaned.groupby(['City', 'Year'])['AveragePrice'].mean().reset_index()
# Get data for earliest and latest year
start_year = yearly_avg['Year'].min()
end_year = yearly_avg['Year'].max()

start_prices = yearly_avg[yearly_avg['Year'] == start_year][['City', 'AveragePrice']].rename(columns={'AveragePrice': 'StartPrice'})
end_prices = yearly_avg[yearly_avg['Year'] == end_year][['City', 'AveragePrice']].rename(columns={'AveragePrice': 'EndPrice'})

# Merge the two sets
price_change = pd.merge(start_prices, end_prices, on='City')
price_change['PriceIncrease'] = price_change['EndPrice'] - price_change['StartPrice']
price_change['%Increase'] = (price_change['PriceIncrease'] / price_change['StartPrice']) * 100
# Sort by highest price increase
top_boroughs = price_change.sort_values(by='PriceIncrease', ascending=False)

# Display top 5
print(top_boroughs[['City', 'StartPrice', 'EndPrice', 'PriceIncrease', '%Increase']].head())
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(borough_data):
    """Takes a borough's yearly data and returns 2018/1998 price ratio."""
    # Get average prices for 1998 and 2018
    price_1998 = borough_data[borough_data['Year'] == 1998]['AveragePrice'].mean()
    price_2018 = borough_data[borough_data['Year'] == 2018]['AveragePrice'].mean()
    
    # Avoid divide-by-zero
    if pd.isna(price_1998) or pd.isna(price_2018) or price_1998 == 0:
        return None
    
    return price_2018 / price_1998
# We'll store results in a dictionary
ratios = {}

# Get unique boroughs
boroughs = yearly_avg['City'].unique()

# Loop through each borough, apply the function
for borough in boroughs:
    borough_data = yearly_avg[yearly_avg['City'] == borough]
    ratio = create_price_ratio(borough_data)
    if ratio:  # Only keep valid results
        ratios[borough] = ratio
ratios_df = pd.DataFrame(list(ratios.items()), columns=['Borough', '2018/1998 Ratio'])
ratios_df.sort_values(by='2018/1998 Ratio', ascending=False, inplace=True)
print(ratios_df.head())
```
